```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , OneHotEncoder 
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report , confusion_matrix ,accuracy_score
```

```python
df=pd.read_csv('dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv')
df
```

```python
df.drop('EmployeeCount' , axis=1 , inplace=True)
```

```python
df.drop('EmployeeNumber' , axis=1 , inplace=True)
```

```python
df.drop('Over18' , axis=1 , inplace=True)
```

```python
df.drop('StandardHours' , axis=1 , inplace=True)
```

```python
x=df.drop('Attrition' , axis=1)
y=df['Attrition']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
num=x.select_dtypes(include=np.number).columns.tolist()
cat=x.select_dtypes(exclude=np.number).columns.tolist()
```

```python
preprocessor=ColumnTransformer(
    transformers=[
        ('num' , StandardScaler() , num),
        ('cat' , OneHotEncoder(handle_unknown='ignore') , cat)
    ]
)
```

```python
pipe=Pipeline(
    steps=[
        ('preprocessor' , preprocessor),
        ('classifier' , LogisticRegression())
    ]
)
```

```python
pipe.fit(x_train , y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
acc=accuracy_score(y_test , y_pred)
cm=confusion_matrix(y_test , y_pred)
class_report=classification_report(y_test , y_pred)

print('Accuracy Score : ' , acc)
print('Confusion Matrix : \n' , cm)
print('Classification Report : \n' , class_report)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_unlimited = Pipeline([
    ('preprocessor', preprocessor),
    ('model', DecisionTreeClassifier(random_state=42))
])

dt_unlimited.fit(x_train, y_train)
```

```python
train_pred = dt_unlimited.predict(x_train)
test_pred = dt_unlimited.predict(x_test)

print("Train Accuracy:",
      accuracy_score(y_train, train_pred))

print("Test Accuracy:",
      accuracy_score(y_test, test_pred))
```

```python
dt_limited = Pipeline([
    ('preprocessor', preprocessor),
    ('model', DecisionTreeClassifier(
        max_depth=5,
        min_samples_split=10,
        random_state=42
    ))
])

dt_limited.fit(x_train, y_train)

```

```python
train_pred = dt_limited.predict(x_train)
test_pred = dt_limited.predict(x_test)

print("Train Accuracy:",
      accuracy_score(y_train, train_pred))

print("Test Accuracy:",
      accuracy_score(y_test, test_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(
        n_estimators=100,
        random_state=42
    ))
])

rf.fit(x_train, y_train)
```

```python
train_pred = rf.predict(x_train)
test_pred = rf.predict(x_test)

print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Test Accuracy:", accuracy_score(y_test, test_pred))
```

```python
from sklearn.ensemble import VotingClassifier

voting_hard = Pipeline([
    ('preprocessor', preprocessor),
    ('model', VotingClassifier(
        estimators=[
            ('lr', LogisticRegression(max_iter=1000)),
            ('dt', DecisionTreeClassifier(max_depth=5)),
            ('rf', RandomForestClassifier(n_estimators=100, random_state=42))
        ],
        voting='hard'
    ))
])

voting_hard.fit(x_train, y_train)
```

```python
train_pred = voting_hard.predict(x_train)
test_pred = voting_hard.predict(x_test)

print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Test Accuracy:", accuracy_score(y_test, test_pred))
```

```python
voting_soft = Pipeline([
    ('preprocessor', preprocessor),
    ('model', VotingClassifier(
        estimators=[
            ('lr', LogisticRegression(max_iter=1000)),
            ('dt', DecisionTreeClassifier(max_depth=5)),
            ('rf', RandomForestClassifier(n_estimators=100, random_state=42))
        ],
        voting='soft'
    ))
])

voting_soft.fit(x_train, y_train)
```

```python
train_pred = voting_soft.predict(x_train)
test_pred = voting_soft.predict(x_test)

print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Test Accuracy:", accuracy_score(y_test, test_pred))
```

```python
from sklearn.ensemble import BaggingClassifier

bag_false = Pipeline([
    ('preprocessor', preprocessor),
    ('model', BaggingClassifier(
        estimator=DecisionTreeClassifier(),
        n_estimators=100,
        bootstrap=False,
        random_state=42
    ))
])

bag_false.fit(x_train, y_train)
```

```python
train_pred = bag_false.predict(x_train)
test_pred = bag_false.predict(x_test)

print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Test Accuracy:", accuracy_score(y_test, test_pred))
```

```python
bag_true = Pipeline([
    ('preprocessor', preprocessor),
    ('model', BaggingClassifier(
        estimator=DecisionTreeClassifier(),
        n_estimators=100,
        bootstrap=True,
        random_state=42
    ))
])

bag_true.fit(x_train, y_train)
```

```python
train_pred = bag_true.predict(x_train)
test_pred = bag_true.predict(x_test)

print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Test Accuracy:", accuracy_score(y_test, test_pred))
```

```python
bag_oob = Pipeline([
    ('preprocessor', preprocessor),
    ('model', BaggingClassifier(
        estimator=DecisionTreeClassifier(),
        n_estimators=100,
        bootstrap=True,
        oob_score=True,
        random_state=42
    ))
])

bag_oob.fit(x_train, y_train)
```

```python
train_pred = bag_oob.predict(x_train)
test_pred = bag_oob.predict(x_test)

print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Test Accuracy:", accuracy_score(y_test, test_pred))

print("OOB Score:", bag_oob.named_steps['model'].oob_score_)
```
